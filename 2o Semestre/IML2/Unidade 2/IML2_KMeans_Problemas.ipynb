# Problemas com algoritmo K-Means #

```python
#importar o pacote datasets do scikit-learn para carregar conjuntos de dados
from sklearn import datasets

#importar KMeans para fazer o agrupamento
from sklearn.cluster import KMeans

#importar o módulo pyplot para plotar o gráfico no método do cotovelo
import matplotlib.pyplot as plt
```

```python
#gerando dados 2D com dois grupos de tamanhos diferentes

X, y = datasets.make_blobs(n_samples=[300,10],centers=[[-2,-1],[2,0]])
plt.scatter(X[:,0], X[:,1], c=y)
plt.title("Grupos com tamanhos muito diferentes")
plt.xlabel("x1")
plt.ylabel("y1")
```

```python
# agrupando com KMeans
kmeans = KMeans(n_clusters = 2, init = 'random')
kmeans.fit(X)
kmeans.cluster_centers_
```

```python
# plotando resultado do KMeans
y_pred = kmeans.labels_
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.title("Grupos encontrados com KMeans")
plt.xlabel("x1")
plt.ylabel("y1")
```

```python
#gerando dados 2D com dois grupos com densidades diferentes

X1, y1 = datasets.make_blobs(n_samples=[150,150],centers=[[-1,1],[5,5]], cluster_std=[1,4])
plt.scatter(X1[:,0], X1[:,1], c=y1)
plt.title("Grupos com densidades muito diferentes")
plt.xlabel("x1")
plt.ylabel("y1")
```

```python
# agrupando com KMeans
kmeans = KMeans(n_clusters = 2, init = 'random')
kmeans.fit(X1)
kmeans.cluster_centers_
```

```python
# plotando resultado do KMeans
y_pred1 = kmeans.labels_
plt.scatter(X1[:,0], X1[:,1], c=y_pred1)
plt.title("Grupos encontrados com KMeans")
plt.xlabel("x1")
plt.ylabel("y1")
#print(plt)
```

```python
#gerando dados 2D com 3 grupos

X2, y2 = datasets.make_blobs(n_samples=[100,200,50])
plt.scatter(X2[:,0], X2[:,1], c=y2)
```

```python
# agrupando com KMeans com inicialização randômica
kmeans = KMeans(n_clusters = 3, init = 'random')
kmeans.fit(X2)
kmeans.cluster_centers_
```

```python
# plotando resultado do KMeans com inicialização randômica
y_pred2 = kmeans.labels_
centroids=kmeans.cluster_centers_
plt.scatter(X2[:,0], X2[:,1], c=y_pred2)
plt.scatter(centroids[:, 0], centroids[:, 1], marker="o", s=169, linewidths=2,
         color="r", zorder=10)
```

```python
# agrupando com KMeans com inicialização com KMeans++
kmeans = KMeans(n_clusters = 3, init = 'k-means++')
kmeans.fit(X2)
kmeans.cluster_centers_
```

```python
# plotando resultado do KMeans com inicialização com KMeans++
y_pred3 = kmeans.labels_
plt.scatter(X2[:,0], X2[:,1], c=y_pred3)
plt.scatter(centroids[:, 0], centroids[:, 1], marker="o", s=169, linewidths=2,
         color="r", zorder=10)
```
